```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import itertools
import os
os.chdir('/home/emi/unipd/Sartori_CBSD/project/cbsdproject')
import text_processing as tp
import topic_analysis as ta
```

```python
PARTY_NAMES = ['Undefined', 'Lega-Salvini', 'Partito Democratico',
               'Fratelli d\'Italia', 'Movimento 5 Stelle', 'Forza Italia ']
```

```python
for party in np.arange(1,6):
    tweets = pd.read_csv(f'data/tweets_party{party}_cleaned.csv')
    texts = tweets['full_text_cleaned']
    texts = texts.str.split()
    clean_words = list(itertools.chain(*texts))
    most_common_words=ta.calculate_words_frecuencies(clean_words)
    plot = ta.plot_words_frecuencies(most_common_words) 
    plot.set_title(f'Most common words - Party {PARTY_NAMES[party]}', fontsize=16)
    fig = plot.get_figure()
    fig.savefig(f'figures/most_common_words_party{party}.png', bbox_inches='tight')
```
